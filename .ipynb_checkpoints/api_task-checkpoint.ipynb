```python
import requests
from concurrent.futures import ThreadPoolExecutor
```

```python
def getNumDraws(year):

    # Write your code here
  
    draft_matches_counter = 0
    
    params = {"year": year}
       
    url = "https://jsonmock.hackerrank.com/api/football_matches"
    
    res = requests.get(url, params=params)
    res_dict = res.json()

    total_pages = res_dict["total_pages"]
    print(f'Total pages: {total_pages}')

    
    page_params = (page for page in range(1, total_pages+1))

    
    def process_page(page_params):
        draw_matches_on_page = 0
        temp_params_dict = {"year": 2011, "page": page_params}
        res = requests.get(url, params=temp_params_dict)
        res_dict = res.json()
        
        for match_data in res_dict["data"]:
            if match_data["team1goals"] == match_data["team2goals"]:
                draw_matches_on_page += 1
                
        return draw_matches_on_page
                 
    # multithreaded page request
    with ThreadPoolExecutor() as pool:
        out = pool.map(process_page, page_params)
        
    total_draw_matches = sum(out)
    print(total_draw_matches)
    
        
    return total_draw_matches
```

```python
getNumDraws(2011)
```
